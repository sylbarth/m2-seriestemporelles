```python
from datetime import date, timedelta

def get_symbol(symbol: str, years: int = 10):
    start_date = date.today() - timedelta(years*365)
    end_date = date.today()
    return pd.DataFrame(yf.download(symbol, start=start_date, end=end_date))     

x = get_symbol("EURUSD=X")
print(x)
x.plot()
```
